```python
vetor = np.arange(1, 11).reshape((-1, 1)) #metodo simplificado para conversao sem precisar saber a quantidade de elementos do vetor.
```

```python
vetor
```

```python
vetor2 = np.arange(10, 21).reshape(-1, 1)
```

```python
vetor2
```

```python
v = np.append(vetor, vetor2).reshape(-1, 1) #faz a juncao dos vetores com shape em colunas( if ausencia .reshape((-1, 1)), vetores retornam a linha
```

```python
v = np.insert(v, 0, 0).reshape((-1, 1)) #acresenta um elemento variavel, posicao para add, elemento a ser adicionado.
```

```python
v
```

```python
np.delete(v, [0]).reshape((-1, 1))
```

```python
n = 100
matrix = np.random.randint(low = 50, high = 2500, size = (n, 5))#para criar uma matriz
colunas = list('ABCDE')#['A', 'B', 'C', 'D', 'E']metodo comum, porem o usado e mais clean
```

```python
df = pd.DataFrame(matrix, columns = colunas, index = range(1, n+1))
```

```python
df.head(10)
```

```python
plt.figure(figsize = (12, 4))#controla as dimensoes da figura.
df['A'].plot(color = 'orange', lw = 1, marker = 'o', markersize = 3) #controla cor, stilo, marcacoes tamanho da linha do grafico
plt.xlabel('Pontuacao')# controla nomeclatura do eixo x
plt.ylabel('Mês')#controla a nomeclatura do eixo y
plt.title('consistência do jogador', fontsize = (20), pad = 20)# controla nomeclatura do grafico, pad = espaco entre o titulo e o grafcio,
plt.grid(alpha = 0.5, linestyle = '--'); #exibe linha de fundo marcadores e controla suas configuracoes 

```
