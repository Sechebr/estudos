# **loc e iloc**

```python
import pandas as pd
```

```python
dados = {'A' : [12, 15, 17, 19],
         'B' : [11, 23, 34, 4],
         'C' : [90, 22, 33, 41]
        }
df = pd.DataFrame(dados, index = ['a', 'b', 'c', 'd'])
```

```python
df
```

```python
df.loc['a'] #filtra o data frame com a linha selecionada(rotulo)
```

```python
df.loc['a' : 'b']
```

```python
df.loc['a' : 'b', 'A'] #filtra a selecao de vetores por limites de colunas
```

```python
df.loc['a' : 'b', 'A' : 'B']
```

```python
df['A'] >= 15 #conceito de selecao booleana.
```

```python
df.loc[(df['A']>=17) & (df['B']>=34)] #selecoes baseadas em condicoes.
```

```python
df.loc[(df['A']>=17) | (df['B']>=34)]
```

```python
df
```

```python
df.iloc[0]
```

```python
df.iloc[0:3]
```

```python
df.iloc[1, 2] #filtro de [linhasxcolunas]
```

```python
df.iloc[1, :2]
```

# selecao com iat/**at**

df.iat[]: acessa um unico valor para um dado par linha / coluna por meio de indexacao

df.at[]: acessa um unico valor para um dado par de rotulos de linha / coluna

```python
import pandas as pd
```

```python
dados = {'A': pd.Series([15, 17, 12], index = ['W', 'K', 'Z']),
         'B': pd.Series([12, 21, 29, 33], index = ['W', 'K', 'Z', 'Y']),
         'C': pd.Series([1, 14, 27, 33], index = ['W', 'K', 'Z', 'Y']),
         'D': pd.Series([101, 22, 33], index = ['K', 'Z', 'Y'])}
df = pd.DataFrame(dados)
```

```python
df
```

```python
df.at['W', 'A'] #acessando pelo rotulo, 
```

```python
df.iat[1,0]
```

```python
#localizando valor 33, na coluna C:
df.at['Y', 'C']
df.iat[2,2] # especicando o indice da linha e da coluna comecando com o 0
```

```python
#substituindo valo com at
df.at['Y', 'A'] = 21
```

```python
df
```

```python
#substituindo valor com iat
df.iat[1, 3] = 130
```

```python
df
```
